# Importing Libraries & Loading DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

## Cleaning Data - Converting the job_posted_date string into datetime

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Filtering the dataframe by the country usa

```python
df_usa = df[df['job_country']=='United States'].copy()
```

```python
df_usa
```

### Extracting the month (in the form of word) value in the new dataframe by creating a new column

```python
df_usa['job_posted_month']=df_usa['job_posted_date'].dt.strftime('%B')
```

```python
df_usa
```

### Pivotting df_usa

```python
df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

### Resetting the index

```python
df_usa_pivot = df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_usa_pivot.reset_index(inplace=True)
```

```python
df_usa_pivot
```

### Adding a month number column

```python


df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot.sort_values('month_no', inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)
```

### Dropping the month_no

```python
df_usa_pivot.drop(columns='month_no', inplace=True)
```

```python
df_usa_pivot
```

### Ploting the dataframe now

```python
df_usa_pivot.plot(kind='line')
```

### Top 3

```python
top_3 = df_usa['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_usa_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Posting for Top Data Jobs in the USA')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
