# Matplotlib Pie & Scatter Plots

## Importing libraries & Loading DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
df = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

## Data Cleanup

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x) # This transforms the string into a list  
```

## Plotting it

## Job work from home

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.2f%%')
plt.title('Work from Home Status')
plt.ylabel('')
plt.show()
```

## Job avaible for home

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job title in the Dataset')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='barh')
plt.title('Work from Home Status')
plt.ylabel('')
plt.gca().invert_yaxis() # This invert the yaxix on horizontal barchart
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(df['job_title_short'].value_counts(), labels=df['job_title_short'].value_counts().index, autopct='%1.2f%%')
ax.set_title('Job Title in the Dataset')
```

## Creating subplots

```python
fig, ax = plt.subplots(1,3) # One row and 3 columns
dict_column = {
  'job_work_from_home': 'Work From Home',
  'job_no_degree_mention': 'Job Degree Req',
  'job_health_insurance': 'Health Insurance offered'
}

for i, (column, title) in enumerate(dict_column.items()):
  #print(i, column, title)
  ax[i].pie(df[column].value_counts(), labels=df[column].value_counts().index,startangle=200, autopct='%1.2f%%')
  ax[i].set_title(title)
plt.tight_layout() # Separates the plots
plt.show()
```
