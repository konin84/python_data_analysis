# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
df = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

## Cleaning Data - Converting the job_posted_date string into datetime

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Selecting the first 10 salary_year_avg  & salary_hour_avg

```python
df.loc[:10,'salary_year_avg':'salary_hour_avg']
```

## Getting the median of salary_year_avg & salary_hour_avg

```python
salary_year_avg_median = df['salary_year_avg'].median()
```

```python
salary_year_avg_median
```

## salary_hour_avg median

```python
salary_hour_avg_median = df['salary_hour_avg'].median()
```

```python
salary_hour_avg_median
```

## Filling NaN values with something else

```python
df_filled =df
```

### Working now with the df_filled

## 1. Filling salary_year_avg

```python
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(salary_year_avg_median)

df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(salary_hour_avg_median)
```

```python
df_filled.loc[:10,'salary_year_avg':'salary_hour_avg']
```

## 2. Dropping dupplicate values

```python
df_unique = df_filled
df_unique=df_unique.drop_duplicates()

print('Length of original df:', len(df))
print('Length of df_unique:', len(df_unique))
print('Length of df_filled:', len(df_filled))
print('Rows Dropped:',  len(df_filled) - len(df_unique))

```
