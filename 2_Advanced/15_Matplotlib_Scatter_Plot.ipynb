# Matplotlib Scatter Plots

## Importing libraries & Loading DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
df = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

## Data Cleanup

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x) # This transforms the string into a list  
```

## Working with job title short of Data Analyst

```python
df_da = df[df['job_title_short'] == 'Data Analyst'].copy()
df_da.head()
```

## Exploding the job skills

```python
df_exploded = df_da.explode('job_skills')
```

## Grouping and aggregating each skill

```python
skill_stats = df_exploded.groupby('job_skills').agg(skill_count=('job_skills', 'count'), 
 median_salary=('salary_year_avg', 'median'))
skill_stats = skill_stats.sort_values('skill_count', ascending=False).head(10)
skill_stats                 
```

## Grouping job skills

```python
skill_stats.plot(kind='scatter', x='skill_count', y='median_salary')
plt.title('Relationship between Skill Count & Median Salary of Top 10 Skills')
plt.xlabel('Skill Count of Jobs Posting')
plt.ylabel('Median Salary ($USD)')
plt.tight_layout()
plt.show()
```

## Labeling the points

```python
skill_stats.plot(kind='scatter', x='skill_count', y='median_salary')
plt.title('Relationship between Skill Count & Median Salary of Top 10 Skills')
plt.xlabel('Skill Count of Jobs Posting')
plt.ylabel('Median Salary ($USD)')
plt.tight_layout()
for i, txt in enumerate(skill_stats.index):
    #plt.annotate(txt, (skill_stats['skill_count'][i], skill_stats['median_salary'][i]))
    plt.text(skill_stats['skill_count'].iloc[i], skill_stats['median_salary'].iloc[i], txt)
plt.show()
```
