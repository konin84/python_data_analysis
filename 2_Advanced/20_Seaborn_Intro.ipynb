```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
df = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

## Data Cleanup

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x) # This transforms the string into a list  
```

## Getting only data analyst jobs in the USA

```python
# Getting only data analyst jobs in the USA
df_da_usa = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] =='United States')].copy()
#df_da_usa['salary_year_avg']
# Droping the NaN values from the column for plotting
df_da_usa = df_da_usa.dropna(subset=['salary_year_avg'])
df_da_usa 
```

## Exploding the job_skills column 

```python
df_da_usa=df_da_usa.explode('job_skills')
#df_da_usa['job_skills']
df_da_usa[['salary_year_avg','job_skills']]
```

```python
df_da_usa_group = df_da_usa.groupby(by='job_skills')['salary_year_avg'].agg(['count', 'median'])

df_da_usa_top_pay = df_da_usa_group.sort_values(by='median', ascending=False).head(10)
df_da_usa_top_skills = df_da_usa_group.sort_values(by='count', ascending=False).head(10).sort_values(by='median', ascending=False).head(10)
df_da_usa_top_skills
```

## Plotting df_da_usa_top_pay & df_da_usa_top_skills 

```python
fig, ax = plt.subplots(2,1) # 2 columns and 1 row

# df_da_usa_top_pay
df_da_usa_top_pay[::-1].plot(kind='barh', y='median', ax=ax[0], legend=False) # [::-1] invert the xaxis
ax[0].set_title('Top 10 Highest Paid skills for Data Analysis')
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : '{:,.0f}K'.format(x/1000)))

# df_da_usa_top_skills
df_da_usa_top_skills[::-1].plot(kind='barh', y='median', ax=ax[1], legend=False)

ax[1].set_title('Top 10 Most In-Demand skills for Data Analysis')
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary (USD)')
#ax[1].xaxis.set_minor_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : '{:,.0f}K'.format(x/1000)))

ax[1].set_xlim(ax[0].get_xlim()) # This allows the second grapth to be of the same axis data
fig.tight_layout() # This removes the overlaps
```

## Using now seaborn for plotting

```python
fig, ax = plt.subplots(2,1) # 2 columns and 1 row

# df_da_usa_top_pay
#df_da_usa_top_pay[::-1].plot(kind='barh', y='median', ax=ax[0], legend=False) # [::-1] invert the xaxis
sns.barplot(data=df_da_usa_top_pay, x='median', y=df_da_usa_top_pay.index, ax=ax[0], hue='median', palette='dark:b_r', legend=False)

ax[0].set_title('Top 10 Highest Paid skills for Data Analysis')
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : '{:,.0f}K'.format(x/1000)))

# df_da_usa_top_skills
#df_da_usa_top_skills[::-1].plot(kind='barh', y='median', ax=ax[1], legend=False)
sns.barplot(data=df_da_usa_top_skills, x='median', y=df_da_usa_top_skills.index, ax=ax[1],  hue='median', palette='light:b', legend=False)

ax[1].set_title('Top 10 Most In-Demand skills for Data Analysis')
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary (USD)')
#ax[1].xaxis.set_minor_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : '{:,.0f}K'.format(x/1000)))

ax[1].set_xlim(ax[0].get_xlim()) # This allows the second grapth to be of the same axis data
fig.tight_layout() # This removes the overlaps
```

## Another plot

```python
# Plotting the histogram

sns.displot(df_da_usa['salary_year_avg'], kind='kde', fill=True)
plt.title('Distribution of the USA Data Analyst yearly salary')
plt.xlabel('Yearly Salary ($USD)')
plt.ylabel('Number of Jobs')

# Modifying the figure
ax=plt.gca()
## Formating the x-axis
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : '{:,.0f}K'.format(x/1000)))
plt.show()
```

## Box Ploting

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
# Filtering the data to appropriate job_titles
df_usa = df[(df['job_title_short'].isin(job_titles)) & (df['job_country'] =='United States')].copy()
# Dropping NaN values of the column salary_year_avg
df_usa = df_usa.dropna(subset=['salary_year_avg'])

job_list = [df_usa[df_usa['job_title_short']==job_title]['salary_year_avg'] for job_title  in job_titles]

sns.boxplot(data=df_usa, x='salary_year_avg', y='job_title_short')
plt.title('Salary Distribution by Job Title in the USA')
plt.xlabel('Yearly Salary ($USD)')
plt.ylabel('Job title short')
plt.xlim(0, 650000)
ticks_x = plt.FuncFormatter(plt.FuncFormatter(lambda x, pos : '${:,.0f}K'.format(x/1000)))

plt.gca().xaxis.set_major_formatter(ticks_x)
plt.show()
```
