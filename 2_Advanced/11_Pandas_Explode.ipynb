# Using the Explode methode in pandas

## Importing libraries & Loading DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
df = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

## Data Cleanup

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x) # This transforms the string into a list  
```

```python
df.head()
```

```python
df[['job_title_short', 'job_skills']] 
```

## Applying explode to job_skills

```python
df_exploded = df.explode('job_skills')
df_exploded
```

```python
df_exploded['job_skills'].value_counts().head(5).plot(kind='bar')
plt.show()
```

```python
skills_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()
skills_count
```

```python
type(skills_count)
```

### Converting the series into data frame

```python
df_skills_count = skills_count.reset_index()
df_skills_count
```

### Assigning a name to the skill counted

```python
df_skills_count = skills_count.reset_index(name='skill_count')
df_skills_count
```

### Sorting the values skill_count

```python
df_skills_count = df_skills_count.sort_values('skill_count', ascending=False)
df_skills_count
```

### Filtering the new data frame

```python
job_title = 'Data Analyst'
top_skills = 15

df_skill_final= df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)
df_skill_final
```

### Ploting the result

```python
df_skill_final.plot(kind='barh', x='job_skills', y='skill_count')
```

### Inverting the display

```python
df_skill_final.plot(kind='barh', x='job_skills', y='skill_count')
plt.gca().invert_yaxis()
plt.title(f"Top {top_skills} Skills for {job_title}'s")
plt.ylabel('')
plt.xlabel('Skill Count')
plt.legend().set_visible(False)
plt.show()
```
