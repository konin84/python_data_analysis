# Pivot_Table

# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
df = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

## Cleaning Data - Converting the job_posted_date string into datetime

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Pandas: Pivot Tables
### df.pivot_table()

### Counts of Job Titles

```python
df.pivot_table(values='salary_year_avg' ,index='job_country', columns='job_title_short' ,aggfunc='median')
```

## Finding the top six countries with the highest median salary_year_avg

```python
top_six_countries = df['job_country'].value_counts().head(6).index
top_six_countries
```

```python
df_job_country_salary = df.pivot_table(values='salary_year_avg' ,index='job_country', columns='job_title_short' ,aggfunc='median')
df_job_country_salary = df_job_country_salary.loc[top_six_countries]

df_job_country_salary
```

#### Filtering by job_title

```python
job_title = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_country_salary = df_job_country_salary[job_title]
df_job_country_salary
```

### Ploting the result

```python
df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary in ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show
```
