# Merging DataFrames

# Importing Libraries & Loading DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
df_software_jobs = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/software_jobs.csv')

df_data_jobs = pd.read_csv('C:/Users/DELL/Desktop/SQL/Data_analysis/data_jobs.csv')
```

```python
df_data_jobs['job_posted_date'] = pd.to_datetime(df_data_jobs['job_posted_date'])
```

```python
df_usa = df_data_jobs[df_data_jobs['job_country']=='United States'].copy()
```

```python
df_usa['job_posted_month']=df_usa['job_posted_date'].dt.strftime('%B')
```

```python
df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_software_jobs
```

## Making the job_posted_month the index

```python
df_software_jobs = df_software_jobs.set_index('job_posted_month')
df_software_jobs
```

## Merging the two data frames

```python
df_usa_pivot = df_usa.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_usa_pivot
```

```python
df_usa_merged = df_usa_pivot.merge(df_software_jobs, on='job_posted_month', how='inner')
df_usa_merged
```

## Getting the top 5

```python
top_5 = (df_usa_merged
         .sum() # Sum the values
         .sort_values(ascending=False) # Sort the values
         .head(5) # Get the top 5
         .index # Get the index of the top 5
         .to_list()) #
top_5
```

## Plotting the obtained Data now

```python
df_usa_merged[top_5].plot(kind='line')
plt.title('Monthly Job Posting for Top Data Jobs in the USA')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 22000)
plt.legend()
plt.show()
```
